```python

import numpy as np
import json
def get_yield_alice_all(outdirectory = "processed"):
    directory = "/home/lieuwe/Documents/Software/articles/1910.07678/hepdata"

    tables = ["Table1", "Table3", "Table5"]


   
    # ranges = [(269, 327),(249, 302),(241,292)]

    particle = [211, 321, 2212]

   
    total = {}
    names = {211 : "chpions_1910.07678", 321 : "chkaons_1910.07678", 2212 : "chprotons_1910.07678"}
    filenames = {211: f"{outdirectory}/dNptdptdy_chpions.json",321: f"{outdirectory}/dNptdptdy_chkaons.json",2212: f"{outdirectory}/dNptdptdy_chprotons.json"}

    for i,p in enumerate([211, 321, 2212]):
        table = tables[i]
        file = open(f"{directory}/{table}.csv", "r")
        lines = file.readlines()
        file.close()
        cin = -1

        data = [ [] for i in range(10)]
        # frange = ranges[i]
        for line in lines:
            # print(line)
            # print(line.find("CENTRALITY"))
            if line.find("CENTRALITY") >0:
                cin += 1
                # print(cin)
            if line.find("#") < 0 and line.find("$") < 0 and len(line) > 1:
                sline = line.strip().split(',')
                
                data[cin].append(sline)
                # print(sline)
        
        data = np.array(data).astype(float)
        
        # print(data)
        mp     = data[:,:,3]/(data[:,:,0] * 2 * np.pi)
        mp_err = (data[:,:,4] + data[:,:,6])/(data[:,:,0] * 2 * np.pi)
        pt = data[:,:,0]
        pt_err = (data[:,:,2] - data[:,:,1])/2




        
        nbinsx = 9
        nbinsy = len(data[:,:,3][0])
        nbinsz = 1

        edgesx = np.array([0, 5, 10, 20, 30, 40, 50, 60, 80, 90])
        
        print(data[:,:,1][0])
        print(data[:,:,2][0][-1])
        edgesy = np.concatenate([data[:,:,1][0], [data[:,:,2][0][-1]]])
        edgesz = np.array([-0.5, 0.5])

        midsx = (edgesx[1:] + edgesx[:-1])/2
        midsy = (edgesy[1:] + edgesy[:-1])/2
        midsz = (edgesz[1:] + edgesz[:-1])/2

        widsx = (edgesx[1:] - edgesx[:-1])
        widsy = (edgesy[1:] - edgesy[:-1])
        widsz = (edgesz[1:] - edgesz[:-1])

        errsx = widsx/2
        errsy = widsy/2
        errsz = widsz/2

    
       
    
        hist = {}

        hist["settings"] = {}
        hist["settings"]["nbins"] = {"x" : int(nbinsx), "y" : int(nbinsy), "z" : int(nbinsz)}
        hist["settings"]["edges"] = {"x" : edgesx.tolist(), "y" : edgesy.tolist(), "z" : edgesz.tolist() }
        hist["settings"]["mids"] =  {"x" : midsx.tolist(), "y" :  midsy.tolist() , "z" : midsz.tolist() }
        hist["settings"]["wids"] =  {"x" : widsx.tolist(), "y" :  widsy.tolist() , "z" : widsz.tolist() }
        hist["settings"]["errs"] =  {"x" : errsx.tolist(), "y" :  errsy.tolist() , "z" : errsz.tolist() }
        hist["settings"]["names"] = {"x" : "centrality", "y" : "transverse momentum", "z" : "rapidity"}
        hist["contents"] = {"vals" : mp.tolist(), "errs" : mp_err.tolist()}
        hist["contents"]["name"] = "d^{2} N / (2 \pi p_{T} dp_{T} dy)"
        hist["name"] = names[p]

        total[p] = hist

    

        with open(filenames[p], "w") as file:
            json.dump(hist, file, indent = 4, separators=(',',':'))




   
    return total
```
